```python
import pandas as pd
df_2019 = pd.read_csv('2019report.csv')
df_2019.head(20)
df_2019.info()
df_2019[df_2019['Healthy life expectancy']=='0.0']
df_2019['Healthy life expectancy'].min()
```

```python
df_2018 = pd.read_csv('2018report.csv')
df_2018['Year'] = 2018
df_2018.head(20)
```

```python
df_2019.head(20)
df_2019['Year'] = 2019
df_2019.head(20)
```

```python
df_total= pd.concat([df_2018, df_2019], axis=0, ignore_index=True)
df_total = df_total.rename(columns={"Country or region": "country", 'Year': 'year'})
df_total.head(20)
```

# Country Table:

```python
df_country_id = df_total['country'].unique()
df_country_id = pd.DataFrame(df_country_id, columns=['country'])
df_country_id['country_id'] = range(1, len(df_country_id) + 1)
df_country_id.head(20)
```

# Year Table:

```python
df_year_id = df_total['year'].unique()
df_year_id = pd.DataFrame(df_year_id, columns=['year'])
df_year_id['year_id'] = range(1, len(df_year_id) + 1)
df_year_id.head(20)
```

```python
df_total2 = df_total.merge(df_country_id, on='country', how='left').drop(columns=['country'])
df_total2
```

```python
df_total3 = df_total2.merge(df_year_id, on='year', how='left').drop(columns=['year'])
df_total3
```

# Metric Table:

```python
metric_columns = df_total3.columns[2:]
df_metrics = df_total3[metric_columns]
df_metrics.info()
df_metrics = df_metrics.rename(columns={"GDP per capita": "gdp_per_capita", "Social support": "social_support", "Healthy life expectancy": "healthy_life_expectancy", "Freedom to make life choices": "freedom_to_make_life_choices", "Generosity": "generosity", "Perceptions of corruption": "perceptions_of_corruption","Country ID": "country_id", "year_id": "year_id"})
df_metrics.head(20)
```

# Score Table:

```python
df_score = df_total3.drop(columns=['GDP per capita','Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption'])
df_score.info()
df_score = df_score.rename(columns={"Score": "score", "Country ID": "country_id", 	'Overall rank': 'overall_rank'})
df_score.head(20)
```

```python
from sqlalchemy import create_engine
engine = create_engine('mysql://root:PASSWORD@localhost/sql_project')
```

```python
df_score.to_sql('score', con=engine, if_exists='replace', index=False)
df_metrics.to_sql('metrics', con=engine, if_exists='replace', index=False)
df_country_id.to_sql('country', con=engine, if_exists='replace', index=False)
df_year_id.to_sql('years', con=engine, if_exists='replace', index=False)
```

```python
my_query = """
SELECT c.country, s2019.overall_rank - s2018.overall_rank AS rank_difference,
m2019.gdp_per_capita - m2018.gdp_per_capita AS capita_difference,
m2019.social_support - m2018.social_support AS support_difference,
m2019.healthy_life_expectancy - m2018.healthy_life_expectancy AS health_difference,
m2019.freedom_to_make_life_choices - m2018.freedom_to_make_life_choices AS freedom_difference,
m2019.generosity - m2018.generosity AS generosity_difference,
m2019.perceptions_of_corruption - m2018.perceptions_of_corruption AS corruption_difference
FROM country AS c
JOIN score AS s2019
ON c.country_id = s2019.country_id
JOIN score AS s2018 
ON c.country_id = s2018.country_id
JOIN metrics AS m2019
ON c.country_id = m2019.country_id
JOIN metrics AS m2018
ON c.country_id = m2018.country_id
WHERE s2018.year_id = 1 AND s2019.year_id = 2 AND m2019.year_id = 2 AND M2018.year_id = 1 AND s2019.overall_rank - s2018.overall_rank = -1
ORDER BY capita_difference DESC;
"""
pd.read_sql_query(my_query, con=engine).head(20)
```

```python
european_countries = [
    'Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic',
    'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece',
    'Hungary', 'Iceland', 'Ireland', 'Italy', 'Latvia', 'Lithuania',
    'Luxembourg', 'Malta', 'Netherlands', 'Norway', 'Poland', 'Portugal',
    'Romania', 'Slovakia', 'Slovenia', 'Spain', 'Sweden', 'Switzerland',
    'United Kingdom'
]

df_2018 = df_total[df_total['year'] == 2018].set_index('country')
df_2019 = df_total[df_total['year'] == 2019].set_index('country')

# Calculate difference
df_diff = df_2019.drop(columns='year') - df_2018.drop(columns='year')

# Keep only European countries
df_diff_europe = df_diff[df_diff.index.isin(european_countries)]

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 10))
sns.heatmap(df_diff_europe, cmap='coolwarm', center=0, annot=True, fmt=".2f")
plt.title("European Countries: Metric Differences (2019 - 2018)")
plt.xlabel("Metrics")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```
